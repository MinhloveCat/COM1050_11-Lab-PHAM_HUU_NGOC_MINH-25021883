```python
#W5A1
def max_of_two(num1, num2):
    if num1 > num2:
        return num1
    else:
        return num2

a,b = map(int, input('Nhập 2 số: ').split())
print(max_of_two(a,b))
```

```python
#W5A2
def swap(two_numbers):
    return two_numbers[::-1]

two_numbers = tuple(map(int, input('Nhập 2 số: ').split()))
print(swap(two_numbers)[0], swap(two_numbers)[1])
```

```python
#5A3
def is_prime(num):
    for i in range(2,int(num*0.5)+1):
        if num % i == 0:
            return False
    return True

a = int(input('Nhập số nguyên dương: '))
if is_prime(a):
    print("True")
else:
    print("False")
```

```python
#W5A5
def find_num_in_lst(num_list, n):
    if n in num_list:
        return num_list.index(n)
    else:
        return -1




num_list = list(map(int, input('Nhập dãy số nguyên: ').split()))
n = int(input('Nhập số nguyên dương k: '))
print(find_num_in_lst(num_list, n))
```

```python
#5A6
def factorial(num):
    if num == 0:
        return 1
    
    fact = 1
    for i in range(1, num+1):
        fact *= i
    return fact

print(factorial(int(input())))
```

```python
#5A7
def calculator(num1, num2, operat):
    if operat == "+":            
        return f"{(num1 + num2):.2f}" 
    elif operat == "-":   
        return f"{(num1 - num2):.2f}" 
    elif operat == "*":
        return f"{(num1 * num2):.2f}" 
    elif operat == "/":
        return f"{(num1 / num2):.2f}" 

num1, operat, num2= input().split()
print(calculator(float(num1), float(num2), operat))
```

```python
#5A8
def hamming_distance(x, y):
    return bin(x ^ y).count('1')

x,y = map(int, input().split())
print(hamming_distance(x,y))
```

```python
#5A9
def tong(num):
    t = 0
    while num > 0:
        t += (num % 10)
        num //= 10
    return t
num = int(input())
print(tong(num))
```

```python
#5A10
def isomorphism(str1, str2):
    mapping_str1_to_str2 = {}
    mapping_str2_to_str1 = {}
    for char_str1, char_str2 in zip(str1, str2):
        print(char_str1, char_str2)
        if char_str1 in mapping_str1_to_str2:
            print(1) #Kiểm tra đã tồn tại ánh xạ chưa
            if mapping_str1_to_str2[char_str1] != char_str2: #Kiểm tra ánh xạ có thỏa mãn ko
                return False
                
        elif char_str2 in mapping_str2_to_str1:
            print(2)
            if mapping_str2_to_str1[char_str2] != char_str1:
                return False
                
        else:
            mapping_str1_to_str2[char_str1] = char_str2
            mapping_str2_to_str1[char_str2] = char_str1
        print(mapping_str1_to_str2, mapping_str2_to_str1)
    return True
str1,str2 = input().split()
print(isomorphism(str1, str2))
```
