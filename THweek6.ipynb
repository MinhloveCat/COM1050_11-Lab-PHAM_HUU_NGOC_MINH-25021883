```python
#W6A9
def add_dict(dict1, dict2):
    for k,v in dict1.items():
        if k in dict2:
            dict2[k] += v
        else:
            dict2[k] = v
    return dict2

```

```python
dict1 = {'Name': 1, 'age': 2}
dict2 = {'name': 5, 'Age': 6}
assert add_dict(dict1, dict2) == {'Age': 6, 'Name': 1, 'age': 2, 'name': 5}
```

```python
#W6A10

def find_pairs_with_sum(lst, k1):
    lst = sorted(lst)
    ans = []
    for i in lst:
        if (k1 - i) in lst:
            if (k1 - i) == i and lst.count(i) == 1: 
                continue
            if ((k1- i ,i) not in ans) and ((i, k1 - i) not in ans):
                ans.append((i, k1 - i))
    ans = sorted(ans, key = lambda x:x[0])
    return ans

        
```

```python


# Test Case 1: Cơ bản (1, 5), (2, 4), (3, 3)
A1 = [1, 5, 2, 4, 3, 3]
k1 = 6
E1 = [(1, 5), (2, 4), (3, 3)]
assert find_pairs_with_sum(A1, k1) == E1, "Test 1 Failed: Basic pairs and sorting"

# Test Case 2: Trùng lặp (3, 3, 3) -> 3 cặp
A2 = [3, 3, 3]
k2 = 6
E2 = [(3, 3)]
assert find_pairs_with_sum(A2, k2) == E2, "Test 2 Failed: Multiple identical pairs"

# Test Case 3: Trùng lặp giá trị (2, 4) xuất hiện 2 lần từ 4 vị trí khác nhau
A3 = [2, 3, 4, 2, 4]
k3 = 6
E3 = [(2, 4)]
assert find_pairs_with_sum(A3, k3) == E3, "Test 3 Failed: Pairs with same value but different indices"

# Test Case 4: Số Âm và Sắp xếp
A4 = [1, 0, -1, 6, 5, 7]
k4 = 6
E4 = [(-1, 7), (0,6), (1,5)]
assert find_pairs_with_sum(A4, k4) == E4, "Test 4 Failed: Negative numbers and proper sorting"

# Test Case 5: Không tìm thấy
A5 = [1, 2, 3, 4, 5]
k5 = 10
E5 = []
assert find_pairs_with_sum(A5, k5) == E5, "Test 5 Failed: No pairs found"

# Test Case 6: List rỗng
A6 = []
k6 = 5
E6 = []
assert find_pairs_with_sum(A6, k6) == E6, "Test 6 Failed: Empty list input"

# Test Case 7: Index Constraint (A[i] + A[j] = k với i <= j)
A7 = [5, 4, 3, 2, 1]
k7 = 6
E7 = [(1, 5), (2, 4)] # A[4]+A[0] bị loại vì 4 > 0. A[1]+A[4] bị loại.
assert find_pairs_with_sum(A7, k7) == E7, "Test 7 Failed: Strict index constraint (i <= j)"
```

```python
#6A11
def split_even_odd(input_tuple):
    odd,even = [], []
    for i in input_tuple:
        if i % 2 == 0:
            even.append(i)
        else: 
            odd.append(i)
    return tuple(even), tuple(odd)
```

```python
input_tuple = tuple(map(int, input().split()))
ans1, ans2 = split_even_odd(input_tuple)
print(f'{ans1}\n{ans2}')
```

```python
#6A12
def find_most_frequent_min(lst):
    result = {}
    for i in lst:
        count = lst.count(i)
        result[i] = count
    if result != {}:
        max_freq = max(result.values())
    else:
        return None
    may_be_min = []
    for i in result:
        if result[i] == max_freq:
            may_be_min.append(i)
    return min(may_be_min)

    
```

```python
lst = list(map(int, input().split()))
print(find_most_frequent_min(lst))
```

```python
#6A13
def reverse_dictionary(lst):
    reversed_lst = {}
    for i in lst:
        reversed_lst[lst[i]] = i
    return reversed_lst
```

```python


# Test Case 1: Cơ bản 1
d1 = {'a': 1, 'b': 2, 'c': 3}
e1 = {1: 'a', 2: 'b', 3: 'c'}
assert reverse_dictionary(d1) == e1, "Test 1 Failed: Basic integer values"

# Test Case 2: Key và Value là các kiểu dữ liệu khác nhau
d2 = {'name': 'Alice', 'age': 30, 'city': 'Hanoi'}
e2 = {'Alice': 'name', 30: 'age', 'Hanoi': 'city'}
assert reverse_dictionary(d2) == e2, "Test 2 Failed: Mixed data types"

# Test Case 3: Giá trị Số âm
d3 = {'d1': -1, 'd2': -2, 'd3': -3}
e3 = {-1: 'd1', -2: 'd2', -3: 'd3'}
assert reverse_dictionary(d3) == e3, "Test 3 Failed: Negative values as new keys"

# Test Case 4: Dictionary rỗng
d4 = {}
e4 = {}
assert reverse_dictionary(d4) == e4, "Test 4 Failed: Empty dictionary"

# Test Case 5: Giá trị là Tuple (hashable)
d5 = {'t1': (1, 2), 't2': (3, 4)}
e5 = {(1, 2): 't1', (3, 4): 't2'}
assert reverse_dictionary(d5) == e5, "Test 5 Failed: Tuple values as new keys"

```

```python
#W6A14
def find_intersection(lst1, lst2):
    result = []
    for i in lst1:
        if i in lst2:
            result.append(i)
    return result
```

```python
#W6A15
def filter_dictionary(dct, k):
    result_dict = {}
    for i in dct:
        if dct[i] > k:
            result_dict[i] = dct[i]
    return result_dict
```

```python
#6A21
def map_string_positions(string):
    string = string.split()
    posi = {}
    for i in string:
        posi[i] = []
    for i in range(len(string)):
        if string[i] in posi:
            posi[string[i]].append(i)
    for i in posi:
        posi[i] = tuple(posi[i])
    return posi
```

```python


input_tc1 = "a b a c b a"
expected_tc1 = {'a': (0, 2, 5), 'b': (1, 4), 'c': (3,)}
assert map_string_positions(input_tc1) == expected_tc1, f"Test Case 1 Thất bại. Expected: {expected_tc1}, Got: {map_string_positions(input_tc1)}"


input_tc2 = "x"
expected_tc2 = {'x': (0,)}
assert map_string_positions(input_tc2) == expected_tc2, f"Test Case 2 Thất bại. Expected: {expected_tc2}, Got: {map_string_positions(input_tc2)}"


input_tc3 = "cat dog cat mouse"
expected_tc3 = {'cat': (0, 2), 'dog': (1,), 'mouse': (3,)}
assert map_string_positions(input_tc3) == expected_tc3, f"Test Case 3 Thất bại. Expected: {expected_tc3}, Got: {map_string_positions(input_tc3)}"


input_tc4 = "1 1 1"
expected_tc4 = {'1': (0, 1, 2)}
assert map_string_positions(input_tc4) == expected_tc4, f"Test Case 4 Thất bại. Expected: {expected_tc4}, Got: {map_string_positions(input_tc4)}"


input_tc5 = "a b c d"
expected_tc5 = {'a': (0,), 'b': (1,), 'c': (2,), 'd': (3,)}
assert map_string_positions(input_tc5) == expected_tc5, f"Test Case 5 Thất bại. Expected: {expected_tc5}, Got: {map_string_positions(input_tc5)}"


input_tc6 = ""
expected_tc6 = {}
assert map_string_positions(input_tc6) == expected_tc6, f"Test Case 6 Thất bại. Expected: {expected_tc6}, Got: {map_string_positions(input_tc6)}"
```
