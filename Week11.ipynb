```python
#Bai_3:
def is_happy_num(arr):
    try:
        sum = 0
        for i in arr:
            sum += int(i)**2
        if sum == 1:
            return 'YES'
        return is_happy_num(list(str(sum)))
    except ValueError:
        return('Loi nhap lieu')
    except RecursionError:
        return 'NO'

num = list(input())
print(is_happy_num(num))
    
    

```

# Regex cơ bản

- `.` : khớp với bất kỳ ký tự nào (trừ xuống dòng). Ví dụ: `a.b` khớp với `acb` hoặc `a1b`.
- `^` : khớp với vị trí **bắt đầu chuỗi**. Ví dụ: `^Hello` khớp với `"Hello world"`.
- `$` : khớp với vị trí **kết thúc chuỗi**. Ví dụ: `world$` khớp với `"Hello world"`.
- `*` : lặp lại ký tự trước đó **0 hoặc nhiều lần**. Ví dụ: `a*` khớp với `""`, `"a"`, `"aa"`, `"aaa"`.
- `+` : lặp lại ký tự trước đó **1 hoặc nhiều lần**. Ví dụ: `a+` khớp với `"a"`, `"aa"`, `"aaa"`.
- `?` : ký tự trước đó có thể **có hoặc không có**. Ví dụ: `a?` khớp với `""` hoặc `"a"`.
- `[]` : biểu diễn một **tập hợp ký tự**. Ví dụ: `[abc]` khớp với `"a"`, `"b"`, hoặc `"c"`.
- `\d` : khớp với một **chữ số** từ 0 đến 9. Ví dụ: `\d{3}` khớp với `"123"`.
- `\w` : khớp với một **ký tự chữ hoặc số** (a-z, A-Z, 0-9, và `_`). Ví dụ: `\w+` khớp với `"abc123"`.
- `\s` : khớp với một **khoảng trắng** (space, tab, xuống dòng). Ví dụ: `\s+` khớp với một hoặc nhiều khoảng trắng liên tiếp.

---

## Ví dụ ứng dụng
- `^\d+$` → kiểm tra chuỗi chỉ toàn số.
- `^[\w\.-]+@[\w\.-]+\.\w+$` → kiểm tra định dạng email.
