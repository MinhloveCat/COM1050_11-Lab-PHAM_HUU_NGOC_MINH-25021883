```python
#W7A1
def binary_search(lst, x):
    left, right = 0, len(lst) - 1
    while left <= right:
        mid = (left + right)//2
        if lst[mid] == x:
            return mid
        elif lst[mid] < x:
            left = mid + 1
        elif lst[mid] > x:
            right = mid - 1
    return -1


```

```python
arr1 = [1, 2 ,2, 3, 3, 4, 5, 6 ,7 ,8]
print(binary_search(arr1, 9))
```

```python
arr2 = [1, 2, 3, 4, 5, 6, 7, 8]
print(binary_search(arr2, 3))
```

```python
#W7A2
def count_occurrences(arr, x):
    count = 0
    for i in arr:
        if i == x:
            count += 1
    return count
```

```python
arr1 = [1 ,2, 2 ,3 ,3, 4 ,5 ,6 ,7, 8 ]
print(count_occurrences(arr1, 9))
```

```python
arr2 = [1,2,3,4,5,6,7,8,3,3]
print(count_occurrences(arr2, 3))
```

```python
#7A3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j + 1] > arr[j]:
                swapped = True
            arr[j],arr[j+1] = arr[j+1],arr[j]
        if not swapped:
            break
    return arr

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1,n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i],arr[min_idx] = arr[min_idx], arr[i]
    
    return arr

```

```python
#W7A4
def occurence(arr):
    freq = {}
    for i in range(len(arr)):
        if arr[i] not in freq:
            freq[arr[i]] = 0
            freq[arr[i]] += 1
        else:
            freq[arr[i]] += 1
    max_freq = max(freq.values())
    candidates = []
    for num in freq:
        if freq[num] == max_freq:
            candidates.append(num)
    result = min(candidates, key = lambda x:arr.index(x))
    return result, max_freq
```

```python
#W7A5
def collection(arr,x):
    pairs = []
    for i in arr:
        if (x - i) in arr:
            if (x - i) == i and frequency(arr)[i] == 1:
                continue
            if (i, (x-i)) not in pairs and ((x-i) ,i) not in pairs:
                pairs.append((i, x-i))
    return pairs
def frequency(arr):
    freq = {}
    for i in range(len(arr)):
        if arr[i] not in freq:
            freq[arr[i]] = 0
            freq[arr[i]] += 1
        else:
            freq[arr[i]] += 1
    return freq
def counter(arr,x):
    sum = 0
    for pair in collection(arr,x):
        sum = sum + (frequency(arr)[pair[0]] * frequency(arr)[pair[1]])
    return sum

arr = [2,3,2,3,1 ,4 ,5, 6 ,-1 ,3, 2 ,3 ]
print(collection(arr,5))
print(frequency(arr))
print(counter(arr,5))
```

```python
#W7A6
def lam_phang_lst(arr):
    result = []
    for sub_lst in arr:
        for i in sub_lst:
            result.append(i)
    return result
```

```python
#W7A7
def suffle(arr):
    cols = []
    cols.append(arr[0])
    for i in range(1,len(arr)):
        if arr[i] > cols[-1]: #Append các phần tử lớn hơn
            cols.append(arr[i])
        elif arr[i] == cols[-1]:
            continue
        else:
            # Dùng binary search để tìm vị trí cho arr[i]
            
            loc = binary_to_find_loc(cols,arr[i])
            if loc != len(cols) - 1:
                continue
            else:
                cols[loc] = arr[i]
            

    # return len(cols)
    return cols


def binary_to_find_loc(lst, x):
    left,right = 0, len(lst) - 1
    while left <= right:
        mid = (left + right) // 2
        if x == lst[mid]:
            return mid
        elif x > lst[mid]:
            left = mid + 1
        else:
            right = mid - 1

    return mid 
    

```
