```python
import math
side = input('Nhập 3 cạnh của một tam giác (cách nhau bằng dấu cách): ').split()
a,b,c = map(float, side)
if a > 0 and b > 0 and c > 0:
  if a + b > c and a + c > b and b + c > a:
    nua_chu_vi = (a+b+c)/2
    area = math.sqrt(nua_chu_vi*abs(nua_chu_vi-a)*abs(nua_chu_vi-b)*abs(nua_chu_vi-c))
    print(f'Diện tích tam giác có độ dài ba cạnh là {a} , {b} , {c} là: {area} ')
  else:
    print('Không phải tam giác')
else:
  print('Không phải tam giác')
```

```python
name = input('Nhập tên của bạn: ')
so_dien_trc = int(input('Nhập chỉ số điện kế tháng trước: '))
so_dien_sau = int(input('Nhập chỉ số điện kế tháng này: '))
so_dien = so_dien_sau - so_dien_trc
def tinh_tien_dien(so_dien, bac_thang):
  tong_tien = 0
  for gioi_han, gia in bac_thang:
    if so_dien <= 0:
      break
    ss_so_dien = min(so_dien, gioi_han)
    tong_tien = tong_tien + ss_so_dien*gia
    so_dien -= ss_so_dien
  return tong_tien

bac_thang = [
    (50, 1984), #Bậc 1
    (50, 2050), #Bậc 2
    (100, 2380), #Bậc 3
    (100, 2998), #Bậc 4
    (100, 3350), #Bậc 5
    (float('inf'), 3460) #Bậc 6

]
print(f'''
Ho va ten: {name}
Tien phai tra la: {round(tinh_tien_dien(so_dien,bac_thang)*1.08)}
''')

```
