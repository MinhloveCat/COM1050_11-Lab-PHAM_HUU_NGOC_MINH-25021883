```python
#W9A1
def card_sort(lst1, lst2):
    if lst1 == []:
        return lst2
    if lst2 == []:
        return lst1
    res = []
    i,j = 0,0
    len_lst1,len_lst2 = len(lst1),len(lst2)
    while True:
        if lst1[i] <= lst2[j]:
            res.append(lst1[i])
            i += 1
            if i == len_lst1:
                res += lst2[j:]
                return res
        else:
            res.append(lst2[j])
            j += 1
            if j == len_lst2:
                res += lst1[i:]
                return res
    
```

```python
lst1 = [1,2,3,4,5]
lst2 = [0,1,1,1,1,1,2,2,3,4,6,7,8,9]
print(card_sort(lst1, lst2))
```

```python
def merge_two_sorted_arr(arr1, arr2):
    if arr1 == []:
        return arr2
    if arr2 == []:
        return arr1
    if arr1[0] <= arr2[0]:
        return [arr1[0]] + merge_two_sorted_arr(arr1[1:],arr2)
    else:
        return [arr2[0]] + merge_two_sorted_arr(arr1,arr2[1:])
```

```python
arr1 = [1,2,3,4,5]
arr2 = [0,1,1,1,1,1,2,2,3,4,6,7,8,9]
print(merge_two_sorted_arr(arr1,arr2))
```

```python
#W9A3
def swap_to_smallest_num(num):
    num_lst = list(str(num))
    min_digit = '9'
    min_idx = -1
    swap_i, swap_j = -1,-1
    for i in reversed(range(len(num_lst))):
        if num_lst[i] < min_digit:
            min_digit = num_lst[i]
            min_idx = i
        if num_lst[i] > min_digit:
            swap_i = i
            swap_j = min_idx
    if swap_i == -1:
        num_lst[-1],num_lst[-2] = num_lst[-2],num_lst[-1]
        return int(''.join(num_lst))
    else:
        num_lst[swap_i],num_lst[swap_j] = num_lst[swap_j], num_lst[swap_i]
        return int(''.join(num_lst))
    

```

```python
num = 123456
new_num = swap_to_smallest_num(num)
print(new_num)
```

```python
#W9A4
def can_transform(arr1, arr2):
    if arr1 == []:
        return "NO"
    if arr2 == []:
        return "YES"
    arr1_pointer,arr2_pointer = 0,0
    if arr1[arr1_pointer] == arr2[arr2_pointer]:
        return can_transform(arr1[arr1_pointer+1:],arr2[arr2_pointer+1:])
    else:
        return can_transform(arr1[arr1_pointer+1:],arr2)




    
```

```python
str1 = ('abCDNCGabcd').lower()
str2 = ('CDNCGABAB').lower()
print(can_transform(list(str1),list(str2)))
```

```python
#9A5
def continuous_subarray(lst1,lst2):
    if lst2 == []:
        return False
    start_idx = lst2.index(lst1[0])
    check = []
    for i in range(start_idx, len(lst2)):
        if lst2[i] not in lst1:
            lst2 = lst2[i+1:]
            if lst1[0] not in lst2:
                return False
            return continuous_subarray(lst1, lst2)
        if lst2[i] not in check:
            check.append(lst2[i])
        if check == lst1:
            return True
    
    
```

```python
lst1 = [1, 2, 3,4]
lst2 = [2,3,1,2,3,5,4,1,6,7,8,1,2,3,5,7,1,1,2,3,5,1,2,3,5,4]

print(continuous_subarray(lst1,lst2))
```

```python
#W9A6
def next_permutation(nums):
    pivot_idx = -1
    for i in range(len(nums)-2,-1,-1):
        if nums[i] < nums[i + 1]:
            pivot_idx = i
            break
    if pivot_idx == -1:
        nums.reverse()
        return nums
    for j in range(len(nums)-1,pivot_idx,-1):
        if nums[j] > nums[i]:
            succesor_idx = j
            break
    nums[pivot_idx],nums[succesor_idx] = nums[succesor_idx],nums[pivot_idx]
    # Reverse the complement
    nums[pivot_idx+1:] = sorted(nums[pivot_idx+1:])
    return nums
```

```python
arr = [1,3,2]
print(next_permutation(arr))
```

```python
#W9A7
def canPlaceFlowers(flowerbed,k):
    
    check_pos = 0
    while check_pos < len(flowerbed):
        if check_pos == 0:
            if flowerbed[check_pos] == 0 and flowerbed[check_pos] == 0:
                k -= 1
                flowerbed[check_pos] = 1
                check_pos += 2
            else:
                check_pos += 2
        elif check_pos == len(flowerbed) - 1:
            if flowerbed[check_pos] == 0 and flowerbed[check_pos - 1] == 0:
                k -= 1
            check_pos += 1
        else:
            if flowerbed[check_pos] == 0:
                if flowerbed[check_pos - 1] != 1 and flowerbed[check_pos + 1] != 1:
                    k -= 1
                    flowerbed[check_pos] = 1
                    check_pos += 2
            elif flowerbed[check_pos] == 1:
                check_pos += 2
        
                
    if k == 0:
        return True
    return False

    




```

```python
#W9A8
class Robot:
    def  __init__(self, x_pos, y_pos):
        self.x_pos  =  x_pos
        self.y_pos = y_pos

    def right(self):
        self.x_pos  += 1
    def left(self):
        self.x_pos -= 1
    def up(self):
        self.y_pos += 1
    def down(self):
        self.y_pos -= 1
    
def is_path_crossing(moves):
    start_pos = (0,0)
    robot = Robot(start_pos[0], start_pos[1])
    log = [start_pos]
    for command in moves:
        if command == "R":
            robot.right()
            if (robot.x_pos,robot.y_pos) not in log:
                log.append((robot.x_pos,robot.y_pos)) 
            else:
                return True
        elif command == "L":
            robot.left()
            if (robot.x_pos,robot.y_pos) not in log:
                log.append((robot.x_pos,robot.y_pos))
            else:
                return True
        elif command == "U":
            robot.up()
            if (robot.x_pos,robot.y_pos) not in log:
                log.append((robot.x_pos,robot.y_pos))
            else:
                return True
        elif command == "D":
            robot.down()
            if (robot.x_pos,robot.y_pos) not in log:
                log.append((robot.x_pos,robot.y_pos))
            else:
                return True
    return False
    
```

```python
commands = 'RURDL'
print(is_path_crossing(commands))
```

```python
#W9A9
def Median(nums):
    a,b,c,d,e = nums
    if a > b:
        a,b=b,a
    if c>d:
        c,d=d,c
    if a>c:
        a,c=c,a
        b,d=d,b
    if e < b:
        return b
    if e > c:
        return c
    return e
nums = list(map(int,input().split()))
print(Median(nums))
```

```python
#9A10
def findRadius(houses,heaters):
    houses.sort()
    heaters.sort()
    i = 0
    r = 0
    for h in houses:
        while i < len(heaters) -1 and abs(heaters[i+1]-h) <= abs(heaters[i]-h):
            i+=1
        r = max(r,abs(heaters[i] - h))
    return r
houses = list(map(int,input().split()))
heaters = list(map(int,input().split()))
print(findRadius(houses,heaters))
```
