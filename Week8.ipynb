```python
class Cylinder:
    
    def __init__(self, radius: float, height:float):
        self.radius = radius
        self.height = height
    
    def surface_area(self):
        bottom_area = (self.radius ** 2) * 3.14
        surrounded_area = (2 * 3.14 * self.radius) * self.height
        return 2 * bottom_area + surrounded_area
    
    def volume(self):
        bottom_area = (self.radius ** 2) * 3.14
        vol = bottom_area * self.height
        return vol
    
```

```python
r,h = map(float, input().split())
cylinder1 = Cylinder(r, h)
s = cylinder1.surface_area()
v = cylinder1.volume()
print(f'{s:.2f}, {v:.2f}')
```

```python
class Date:
    valid_month_and_day = {
            1:31,
            2:28,
            3:31,
            4:30,
            5:31,
            6:30,
            7:31,
            8:31,
            9:30,
            10:31,
            11:30,
            12:31
        }
    def __init__(self, date, month,year):
        self.date = date
        self.month = month
        self.year = year

    def is_leap_year(self):
        return (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 !=0)

    def is_valid(self):
        if self.date < 0 or self.month < 0 or self.year < 0:
            return False
        
        day_in_month = Date.valid_month_and_day.copy()
        
        # Leap_year processing
        if self.is_leap_year():
            day_in_month[2] = 29
        
        # Check valid
        
        if 1 <= self.month <= 12:
            if self.date <= day_in_month[self.month]:
                return True
            
        return False
            
    def next_day(self):
        day_in_month = Date.valid_month_and_day.copy()
        if self.is_leap_year():
            day_in_month[2] = 29
        
        self.date += 1
        if self.date > day_in_month[self.month]:
            self.date = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1
            
        
        
    
    def __str__(self):
            return f"{self.date:02d}/{self.month:02d}/{self.year}"
```

```python
date, month, year = map(int, input().split('/'))
day = Date(date,month,year)
if day.is_valid():
    day.next_day()
    print(day)
else:
    print("INVALID")
```

```python
class Point:
    def __init__(self, x_value, y_value):
        self.x = x_value
        self.y = y_value
    
    def is_origin(self):
        return self.x == 0 and self.y == 0
    
    def on_which_axis(self):
        if self.x == 0:
            return "Nằm trên trục tung"
        if self.y == 0:
            return "Nằm trên trục hoành"
        else:
            return "Không nằm trên trục tọa độ"
    def dist(self):
        return ((self.x)**2 + (self.y)**2)**0.5
```

```python
x,y = map(float, input().split())
point_M = Point(x,y)
if point_M.is_origin():
    print('Là gốc tọa độ')
else:
    print('Không là gốc tọa độ')
print(point_M.on_which_axis())
print(f'Khoảng cách từ M đến gốc tọa độ là {point_M.dist():.2f}')
```

```python
class Calculator:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def add(self):
        print(f"{self.x} + {self.y} = {self.x + self.y}")
    def subtract(self):
        print(f"{self.x} - {self.y} = {self.x - self.y}")
    def product(self):
        print(f"{self.x} * {self.y} = {self.x * self.y}")
    def divide(self):
        print(f"{self.x} / {self.y} = {self.x / self.y}")
    def power(self):
        print(f"{self.x} ^ {self.y} = {self.x ** self.y}")
    def mod(self):
        print(f"{self.x} % {self.y} = {self.x % self.y}")
    def set_numbers(self, new_x, new_y):
        self.x = new_x
        self.y = new_y
```

```python
class ShoppingCart:
    def __init__(self):
        self.cart = []

    def add_product(self):
        sp = input('Nhập sản phẩm: ')
        price = float(input('Nhập giá tiền của sản phẩm: '))
        self.cart.append({sp:price})
    def remove_product(self):
        w_sp = input('Nhập tên sản phẩm: ')
        xoa = False
        for i in self.cart:
            if list(i.keys())[0] == w_sp:
                self.cart.remove(i)
                xoa = True
                break
        if xoa:
            print(f'Đã xóa sản phẩm {w_sp}')
        else:
            print(f'Không tìm thấy sản phẩm {w_sp}')

    def check_cart(self):
        if self.cart != []:
            print('Giỏ hàng không bị rỗng')
        else:
            print('Giỏ hàng rỗng')
    def count_money(self):
        count = 0
        for i in self.cart:
            count += list(i.values())[0]
        print(f'Tổng số tiền của giỏ hàng: {count}')
    def show_cart(self):
        if self.cart:
            for i in self.cart:
                print(i,end = '\n')
        else:
            print(self.cart)
    def delete(self):
        self.cart = []
```

```python
def menu():
    print("\n--- MENU ---") 
    print("1. Thêm một sản phẩm mới") 
    print("2. Xóa 1 sản phẩm khỏi giỏ hàng") 
    print("3. Kiểm tra giỏ hàng") 
    print("4. Tính tiền") 
    print("5. Hiển thị giỏ hàng") 
    print("6. Xóa toàn bộ") 
    
```

```python
stillShopping = True
gio_hang = ShoppingCart()
while stillShopping:
    menu()
    opt = input("Nhập số trước mỗi lựa chọn: ")
    match opt:
        case '1': gio_hang.add_product()
        case '2': gio_hang.remove_product()
        case '3': gio_hang.check_cart()
        case '4': gio_hang.count_money()
        case '5': gio_hang.show_cart()
        case '6': gio_hang.delete()
    cont = input("Bạn có muốn thoát? 'y'/'n': ")
    if cont == 'y':
        stillShopping = False
    
        
```

```python
class Rectangle:
    def __init__(self, w,h):
        self.width = w
        self.height = h
    
    def area(self):
        return self.width * self.height

    def perimeter(self):
        return (self.width + self.height)*2
    
    def scale(self,w,h ,k):
        self.width = w*k
        self.height = h*k
    
```

```python
w,h,k = map(float, input().split())
rect = Rectangle(w,h)
rect.scale(w,h,k)
area,perimeter = rect.area(), rect.perimeter()
print(area, perimeter)
```

```python
import math
class Fraction:
    def __init__(self,num, den): #numerator/denominator
        if den == 0:
            raise ValueError('Nhập mẫu = 0')
        if den < 0:
            den = -den
        cm = math.gcd(num, den)
        self.numerator = num//cm
        self.denominator = den//cm
    
    def add(self,other):
        lcm_den = math.lcm(self.denominator, other.denominator)
        new_num = self.numerator * (lcm_den // self.denominator) + other.numerator * (lcm_den // other.denominator) 
        new_den = lcm_den
        return Fraction(new_num, new_den)
    def sub(self,other):
        lcm_den = math.lcm(self.denominator, other.denominator)
        new_num = self.numerator * (lcm_den // self.denominator) - other.numerator * (lcm_den // other.denominator) 
        new_den = lcm_den
        return Fraction(new_num, new_den)
    def mul(self,other):
        new_num = self.numerator * other.numerator  
        new_den = self.denominator * other.denominator
        return Fraction(new_num, new_den)
    def div(self,other):
        new_num = self.numerator * other.denominator
        new_den = self.denominator * other.numerator
        return Fraction(new_num, new_den)
        

        
    def __str__(self):
        return f"{self.numerator}/{self.denominator}"
```

```python
class BankAccount:
    def __init__(self, name, balance):
        self.owner = name
        self.balance = balance

    def deposit(self, depo):
        self.balance += depo   
        return self            

    def withdraw(self, wdraw):
        if wdraw <= self.balance:
            self.balance -= wdraw
        return self            

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"


# --- Demo ---
acc = BankAccount('Kim', 200)
acc.deposit(0.5).withdraw(50.25)   
print(acc)
```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.avg_score = 0
        self.rank = ''

    def avg(self, score):
        for i in score:
            self.avg_score += i
        self.avg_score /= len(score)
        return self
    def ranking(self):
        if self.avg_score >= 8:
            self.rank = 'Excellent'
        elif self.avg_score >= 6.5:
            self.rank = 'Good'
        elif self.avg_score >= 5:
            self.rank = 'Average'
        else:
            self.rank = 'Poor'
        return self
    
    def __str__(self):
        return f'{self.name} {self.avg_score:.2f} {self.rank}'
```

```python
name = input('Name: ')
n = int(input('n = '))
score = []
for i in range(n):
    point = float(input("Subject score: ").split()[1])
    score.append(point)
student = Student(name)
print(student.avg(score).ranking())
```

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

class Library:
    def __init__(self):
        self.books = []
    def add(self,book):
        book_info = str(book).split(';')
        title,author,year = book_info[0],book_info[1],book_info[2]
        self.books.append(Book(title, author,year))
    def count_by_author(self,name):
        cout = 0
        m = len(self.books)
        for i in range(m):
            if self.books[i].author == name:
                cout += 1
        return cout
    def find_by_year(self,year):
        cout = 0
        m = len(self.books)
        for i in range(m):
            if self.books[i].year == year:
                cout += 1
        return cout
lib = Library()
n = int(input())
for i in range(n):
    opt = input().split()
    if opt[0] == "ADD":
        lib.add(opt[1])
    elif opt[0] == "COUNT":
        print(lib.count_by_author(opt[1]))
    elif opt[0] == "COUNTYEAR":
        print(lib.find_by_year(opt[1]))
    

    
```

```python
#W8A12
class ComplexNumber:
    def __init__(self,real,imagin):
        self.real = real
        self.imagin = imagin
    
    def multiply(self,other):
        new_real = (self.real*other.real) - (self.imagin*other.imagin) 
        new_imagin = (self.real*other.imagin) + (self.imagin*other.real)
        return new_real, new_imagin
    
```

```python
com1_real,com1_imagin = map(int,input().split())
com2_real,com2_imagin = map(int,input().split())    
com1 = ComplexNumber(com1_real,com1_imagin)     
com2 = ComplexNumber(com2_real,com2_imagin)

print(com1.multiply(com2)[0], com1.multiply(com2)[1])
```

```python
#W8A13
class Triangle:
    def  __init__(self,edge1,edge2,edge3):
        self.edge1 = edge1
        self.edge2 = edge2
        self.edge3 = edge3
    def is_triangle(self):
        if self.edge1 + self.edge2 <= self.edge3 or self.edge1 + self.edge3 <= self.edge2 or self.edge2 + self.edge3 <= self.edge1:
            return False
        return True
    def getArea(self):
        half_sur = (self.edge1 + self.edge2 + self.edge3) / 2
        area = (half_sur*abs(half_sur  - self.edge1)*abs(half_sur  - self.edge2)*abs(half_sur  - self.edge3))**0.5
        return area
            
        

edge1, edge2 ,edge3= map(int,input().split())
tri = Triangle(edge1, edge2, edge3)
if tri.is_triangle():
    print(tri.getArea())
else:
    print('invalid')
    
```
