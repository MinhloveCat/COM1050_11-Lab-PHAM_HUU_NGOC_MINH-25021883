```python
#W4A6
import math

# Hàm kiểm tra số nguyên tố
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True



while True:
    try:
        a, b = map(int, input().split())
        if a > b or a <= 0 or b <= 0:
            print("Điều kiện: a và b là số nguyên dương, a ≤ b. Nhập lại!")
            continue

        tong = 0
        for i in range(a, b + 1):
            if is_prime(i):
                tong += i
        print(f"Tổng các số nguyên tố trong đoạn [{a}, {b}] là: {tong}")
    except ValueError:
        print("Vui lòng nhập hai số nguyên hợp lệ!")
```

```python
#W4A7
import math

# Nhập số nguyên dương n
while True:
    try:
        n = int(input("Nhập số nguyên dương n (n ≥ 2): "))
        if n >= 2:
            break
        else:
            print("n phải ≥ 2! Nhập lại.")
    except ValueError:
        print("Vui lòng nhập một số nguyên hợp lệ!")

# Tìm ước số nguyên tố lớn nhất
largest_prime = -1
temp = n

# Chia hết cho 2
while temp % 2 == 0:
    largest_prime = 2
    temp //= 2

# Chia cho các số lẻ từ 3 đến sqrt(n)
for i in range(3, int(math.sqrt(temp)) + 1, 2):
    while temp % i == 0:
        largest_prime = i
        temp //= i

# Nếu sau cùng còn lại một số > 2 thì chính nó là số nguyên tố lớn nhất
if temp > 2:
    largest_prime = temp

print(f"Ước số nguyên tố lớn nhất của {n} là: {largest_prime}")
```

```python
#W4A8
def is_palindrome(n):
    return str(n) == str(n)[::-1]

def reverse_add_palindrome(n):
    steps = 0
    while not is_palindrome(n):
        n += int(str(n)[::-1])
        steps += 1
    return steps, n


n = int(input())
steps, result = reverse_add_palindrome(n)
print(steps, result)
```

```python
#W4A9
import math

def distinct_digits(num):
    s = str(num)
    return len(set(s)) == len(s)

def perfect_squares_distinct_digits(n):
    result = []
    for k in range(1, int(math.isqrt(n)) + 1):
        sq = k * k
        if distinct_digits(sq):
            result.append(str(sq))
    if result:
        print(" ".join(result))
    else:
        print("no number")


n = int(input())
perfect_squares_distinct_digits(n)
```

```python
#W4A10
def collazt(num):
    count = 1
    while num > 1:
        if num % 2 == 0:
            num = num /2
        else:
            num = (num*3) +1
        count +=1
    return count

def main():
    n = int(input("Nhap n: "))
    max = collazt(1)
    num = 1
    for i in range(1, n+1):
        if collazt(i) > max:
            max = collazt(i)
            num = i
    print(num, max)

main()
```

```python
#W4a13
def divisor(num):
    di_list = []
    for i in range(1, int(num**0.5)+1):
        if num % i == 0:
            di_list.append(i)
            di_list.append(num/i)
    return di_list

def is_close_pair(num1, num2):
    sum1, sum2 = 0 , 0
    for i in divisor(num1):
        sum1 += i
    for j in divisor(num2):
        sum2 += j
    if (sum1-num1) == num2 and (sum2-num2) == num1:
        return True
    else:
        return False

def main():
    num1 = int(input('Nhap n: '))
    num2 = int(input('Nhap m: '))
    if is_close_pair(num1, num2):
        print('True')
    else:
        print('False')

main()
    


    

    

```

```python
#w4a32
def count_in_hoa(password):
    in_hoa = 0
    for i in password:
        if 65 <= ord(i) <= 90:
            in_hoa+=1
    return in_hoa

def count_so(password):
    so = 0
    for i in password:
        if 48 <= ord(i) <= 57:
            so+=1
    return so

def count_char(password):
    char = 0
    for i in password:
        if (33 <= ord(i) <= 47) or (58 <= ord(i) <= 64) or (91 <= ord(i) <= 96) or (123 <= ord(i) <= 126):
            char+=1
    return char

def count_thg(password):
    thg = 0
    for i in password:
        if 97 <= ord(i) <= 122:
            thg+=1
    return thg

def check_password(password):
    if len(password) <= 6:
        print('Mật khẩu yếu!')
    else:
        if count_char(password) >= 1 and count_in_hoa(password) >= 1 and count_so(password) >= 1 and count_thg(password) >= 1:
            print('Mật khẩu mạnh!')

def main():
    password = input('Nhập mật khẩu: ')
    check_password(password)

if __name__ == "__main__":
    main()
```
