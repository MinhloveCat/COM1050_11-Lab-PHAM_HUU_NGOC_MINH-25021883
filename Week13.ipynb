```python
#Bai_4:
def is_anagram(path1,path2):
    with open(path1,'r') as file1:
        lst1 = file1.read().lower().split()
    with open(path2,'r') as file2:
        lst2 = file2.read().lower().split()
    str1,str2 = ''.join(lst1),''.join(lst2)
    lst1,lst2 = map(list,[str1,str2])
    dict1,dict2 = {},{}
    for i in lst1:
        if i not in dict1:
            dict1[i] = lst1.count(i)
    for j in lst2:
        if j not in dict2:
            dict2[j] = lst2.count(j)
    
    for k in dict1.keys():
        if k not in dict2:
            return False
        else:
            if dict2[k] != dict1[k]:
                return False
    return True

    
    
```

```python
print(is_anagram('Tom_Marvolo_Riddle.txt','I_Am_Lord_Voldemort.txt'))
```

```python
#Bai_8:
def fib(n):
    if n == 0:
        return 'a'
    if n == 1:
        return 'b'
    return fib(n-1) + fib(n-2)
```

```python
def rob(lst):
    lst = list(map(int,lst))
    if len(lst) < 1:
        return 0
    if len(lst) == 1:
        return lst[0]
    
    max_money = [0] * len(lst)
    max_money[0] = lst[0]
    max_money[1] = max(lst[0],lst[1])
    for i in range(2,len(lst)):
        max_money[i] = max(max_money[i-1], max_money[i - 2] + lst[i])

    return max_money[-1]
    
```

```python
lst = getMoney('HouseRobber.txt')
print(rob(lst))
```

```python
#Bai_13:
def moveZeros(path):
    with open(path,'r') as file:
        lst = file.read().split()
    lst = list(map(int,lst))
    count = 0
    for i in range(len(lst)-1):
        if lst[i] == 0:
            count += 1
            lst.remove(lst[i])
            lst.append(0)
    return lst

with open('dich_chuoi','w') as file:
    file.write('1 0 2 7 0 3')

print(moveZeros('dich_chuoi'))

```

```python
#Bai 19:
import struct

# Nhập đường dẫn tệp từ bàn phím
path = input("Nhập đường dẫn tệp nhị phân: ")

try:
    with open(path, 'rb') as f:
        while True:
            bytes_read = f.read(8)  # mỗi số double chiếm 8 byte
            if not bytes_read:
                break
            number = struct.unpack('d', bytes_read)[0]
            print(f"{round(number, 5)}")
except FileNotFoundError:
    print("Không tìm thấy tệp.")
except Exception as e:
    print("Lỗi:", e)
```

```python
#Bai_18
def estimatedTime(path):
    with open(path,'r') as file:
        data = file.read().strip().split('\n')
    start_point = data[0].split()
    start_point_x,start_point_y = map(int,[start_point[2],start_point[4]])
    target = data[1].split()
    target_x,target_y = map(int,[target[2],target[4]])
    velocity = int(data[2].split()[1])
    dist = ((target_x - start_point_x)**2 + (target_y - start_point_y)**2)**0.5
    time = dist / velocity
    print(f'{time:.2f}')


```

```python
estimatedTime('estimatedTime.txt')
```

```python
import math

def imgSmoother(path):
    with open(path, 'r') as file:
        img = file.read().strip().split('\n')

    # Chuyển thành ma trận số nguyên
    img = [list(map(int, row.split())) for row in img]

    m = len(img)
    n = len(img[0])  # phòng trường hợp ma trận không vuông

    # Tạo ma trận mới để lưu kết quả
    new_img = [[0]*n for _ in range(m)]

    for i in range(m):
        for j in range(n):
            total = 0
            count = 0
            # duyệt qua các ô lân cận (bao gồm chính nó)
            for x in range(i-1, i+2):
                for y in range(j-1, j+2):
                    if 0 <= x < m and 0 <= y < n:
                        total += img[x][y]
                        count += 1
            new_img[i][j] = math.floor(total / count)

    # In kết quả
    for row in new_img:
        print(row)
```
