```python
#Bai_4:
def is_anagram(path1,path2):
    with open(path1,'r') as file1:
        lst1 = file1.read().lower().split()
    with open(path2,'r') as file2:
        lst2 = file2.read().lower().split()
    str1,str2 = ''.join(lst1),''.join(lst2)
    lst1,lst2 = map(list,[str1,str2])
    dict1,dict2 = {},{}
    for i in lst1:
        if i not in dict1:
            dict1[i] = lst1.count(i)
    for j in lst2:
        if j not in dict2:
            dict2[j] = lst2.count(j)
    
    for k in dict1.keys():
        if k not in dict2:
            return False
        else:
            if dict2[k] != dict1[k]:
                return False
    return True

    
    
```

```python
print(is_anagram('Tom_Marvolo_Riddle.txt','I_Am_Lord_Voldemort.txt'))
```

```python
#Bai_8:
def fib(n):
    if n == 0:
        return 'a'
    if n == 1:
        return 'b'
    return fib(n-1) + fib(n-2)
```

```python
with open('output.txt','w') as file:
    n = int(input())
    for i in range(n+1):
        file.write(f'f({i}) = {fib(i)}\n')


```
